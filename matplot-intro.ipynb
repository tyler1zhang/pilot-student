```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, label = 'cos', linestyle = '--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()

plt.show()
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('sine')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('cosine')

plt.show()
```

## The Key of Plot is Axe
[matplot official web](https://matplotlib.org/faq/usage_faq.html)

```python
x = np.arange(0, 20, 0.1)
y1 = x ** 0.5
y2 = np.sin(x)
y3 = np.cos(x)
```

```python
fig, axes = plt.subplots(2, 1, figsize= (14, 12))

axes[0].plot(x, y1, label = 'sqrt')
axes[0].plot(x, y2, label = 'sin')

axes[0].set_title('sqrt(x) vs sin(x)', fontsize = 18)
axes[0].legend()

axes[1].plot(x, y1, label = 'sqrt')
axes[1].plot(x, y3, label = 'cos')

axes[1].set_title('sqrt(x) vs cos(x)', fontsize = 18)
axes[1].legend()

axes[1].set_xlabel('x axis', fontsize = 14, fontstyle = 'italic')
axes[1].set_ylabel('y axis', fontsize = 'x-large', fontstyle = 'oblique')

axes[1].set_aspect('equal')

axes[1].set_xlim(0, 16)

start, end = axes[1].get_xlim()
axes[1].xaxis.set_ticks(np.arange(start, end, 1))

axes[1].xaxis.set_tick_params(rotation = 45, labelsize= 'large', color = 'b')
axes[1].yaxis.tick_right()

axes[1].minorticks_on()

axes[1].grid(which='major', axis = 'both', color ='r')
axes[1].grid(which = 'minor', axis='both')

for ax in axes:
    ax.set_anchor('W')

plt.subplots_adjust(hspace = 0.3)

plt.show()
```
